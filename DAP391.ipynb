```python
from PIL import Image
from diffusers import StableDiffusionControlNetPipeline, ControlNetModel, UniPCMultistepScheduler
import torch
from controlnet_aux import MLSDdetector
from diffusers.utils import load_image
import torch.nn as nn
from PIL import Image
import torch
import torch.nn as nn
import torchvision.transforms as transforms
```

```python
# Tải model ControlNet và pipeline
controlnet = ControlNetModel.from_pretrained(
    "lllyasviel/sd-controlnet-mlsd", torch_dtype=torch.float16
)

pipe = StableDiffusionControlNetPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5", controlnet=controlnet, safety_checker=None, torch_dtype=torch.float16
)

# Tải và cài đặt MLSD detector
mlsd = MLSDdetector.from_pretrained('lllyasviel/ControlNet')
```

```python
def get_image_tensor(image):
    transform = transforms.ToTensor()  # Chuyển ảnh PIL thành tensor
    image_tensor = transform(image).unsqueeze(0)  # Thêm chiều batch
    return image_tensor
```

```python
image = Image.open("/content/drive/MyDrive/Colab Notebooks/Shark0.JPEG")
image_tensor = get_image_tensor(image)
```

```python
#khởi tạo lớp DCL
class DynamicConditioningLayer(nn.Module):
    def __init__(self, style_dim, feature_dim):
        super(DynamicConditioningLayer, self).__init__()
        self.style_transform = nn.Linear(style_dim, feature_dim)

    def forward(self, features, style_vector):
        style_adjustment = self.style_transform(style_vector)
        # Reshape style_adjustment to match the number of features
        style_adjustment = style_adjustment[:, :, None, None].repeat(1, 1, features.shape[2], features.shape[3])
        conditioned_features = features * (1 + style_adjustment)
        return conditioned_features
```

```python
style_dict = {
    "Van Gogh Style": torch.tensor([1.0, 0.5, -0.5]),
    "Artistic Style": torch.tensor([0.3, 0.8, 0.1]),
    "Impressionism": torch.tensor([0.6, 0.7, 0.2]),
    "Cubism": torch.tensor([-0.5, 0.6, 0.9]),
    "Abstract Expressionism": torch.tensor([0.9, -0.3, 0.5]),
    "Surrealism": torch.tensor([0.1, 0.4, 0.8]),
    "Realism": torch.tensor([0.7, 0.2, 0.3]),
    "Pop Art": torch.tensor([-0.2, 0.9, 0.1]),
    "Futurism": torch.tensor([0.8, 0.1, -0.6]),
    "Minimalism": torch.tensor([-0.1, 0.2, 0.9]),
    # Thêm các phong cách khác vào đây
}

# Tạo lớp Dynamic Conditioning Layer
feature_dim = image_tensor.shape[1]  # Lấy số kênh của ảnh tensor
dcl = DynamicConditioningLayer(style_dim=3, feature_dim=feature_dim)
```

```python
def apply_style_condition(image, style_name, pipe, dcl, num_inference_steps=20):
    # Lấy vector phong cách từ style_dict
    style_condition = style_dict.get(style_name, torch.zeros(3))  # Vector mặc định nếu không có phong cách
    style_condition = style_condition.unsqueeze(0)  # Thêm chiều batch

    # Chuyển đổi ảnh từ PIL sang tensor và chuẩn hóa
    transform = transforms.ToTensor()
    image_tensor = transform(image).unsqueeze(0)  # Thêm chiều batch

    # Điều chỉnh các đặc trưng của ảnh bằng DCL
    with torch.no_grad():
        # Get the original shape of the image tensor
        original_shape = image_tensor.shape

        # Reshape to (batch_size, channels, height, width) if necessary
        if len(original_shape) == 2:  # If grayscale
            image_tensor = image_tensor.unsqueeze(1)  # Add channel dimension
            original_shape = image_tensor.shape  # Update original shape


        # image_features = image_tensor.view(1, -1)  # Remove flattening
        conditioned_features = dcl(image_tensor, style_condition)
        # conditioned_image = conditioned_features.view(image_tensor.shape)  # Chuyển về kích thước ban đầu
        conditioned_image = conditioned_features  # No need to reshape

    # Chuyển đổi lại từ tensor sang định dạng mà pipeline chấp nhận
    conditioned_image_pil = transforms.ToPILImage()(conditioned_image.squeeze(0))

    # Truyền qua pipeline để tạo ra ảnh
    with torch.no_grad():
        output = pipe(
            "Shark with " + style_name,
            conditioned_image_pil,
            num_inference_steps=num_inference_steps
        ).images[0]

    return output
```

```python
output_image = apply_style_condition(image, "Abstract Expressionism", pipe, dcl)

# Lưu ảnh kết quả
output_image.save('/content/drive/MyDrive/Colab Notebooks/Shark88.JPEG')
```
